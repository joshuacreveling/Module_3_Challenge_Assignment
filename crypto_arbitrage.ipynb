##### 4.1.0 Use conditional statement to set higher_priced_exchange_early/middle/late and lower_priced_exchange_early/middle/late dataframes to allow for a more dynamic date anlaysis. 

```python
# By conditionally assigning the higher_priced exchanges and lower_price_exchanges, this allows us to optimize profit for the day being analyzed as well as correctly calculating
# arbitrage spread gains.

if bitstamp_sliced.loc[early_date].mean() >= coinbase_sliced.loc[early_date].mean():
    higher_priced_exchange_early = bitstamp_sliced.loc[early_date]
    lower_priced_exchange_early = coinbase_sliced.loc[early_date]
else:
    higher_priced_exchange_early = coinbase_sliced.loc[early_date]
    lower_priced_exchange_early = bitstamp_sliced.loc[early_date]
    
if bitstamp_sliced.loc[middle_date].mean() >= coinbase_sliced.loc[middle_date].mean():
    higher_priced_exchange_middle = bitstamp_sliced.loc[middle_date]
    lower_priced_exchange_middle = coinbase_sliced.loc[middle_date]
else:
    higher_priced_exchange_middle = coinbase_sliced.loc[middle_date]
    lower_priced_exchange_middle = bitstamp_sliced.loc[middle_date]
    
if bitstamp_sliced.loc[late_date].mean() >= coinbase_sliced.loc[late_date].mean():
    higher_priced_exchange_late = bitstamp_sliced.loc[late_date]
    lower_priced_exchange_late = coinbase_sliced.loc[late_date]
else:
    higher_priced_exchange_late = coinbase_sliced.loc[late_date]
    lower_priced_exchange_late = bitstamp_sliced.loc[late_date]
```
